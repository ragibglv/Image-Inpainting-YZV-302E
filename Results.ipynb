```python
import pandas as pd
```

```python
df = pd.read_csv("training_metrics.csv", index_col=0)
df_loss = df[["Train Loss", "Validation Loss"]]
df_psnr = df[["Train PSNR", "Validation PSNR"]]
df_accuracy = df[["Train Accuracy", "Validation Accuracy"]]
df_mse = df[["Train MSE", "Validation MSE"]]
```

```python
df2 = pd.read_csv("training_metrics_unet_with_headers.csv", index_col=0)

df2_loss = df2[["Train Loss", "Validation Loss"]]
df2_psnr = df2[["Train PSNR", "Validation PSNR"]]
df2_accuracy = df2[["Train Accuracy", "Validation Accuracy"]]
df2_mse = df2[["Train MSE", "Validation MSE"]]
```

```python
import matplotlib.pyplot as plt
import numpy as np

epochs_df = np.arange(1, len(df) + 1)
epochs_df2 = np.arange(1, len(df2) + 1)

metrics = {
    "Loss": ("Train Loss", "Validation Loss"),
    "PSNR": ("Train PSNR", "Validation PSNR"),
    "Accuracy": ("Train Accuracy", "Validation Accuracy"),
    "MSE": ("Train MSE", "Validation MSE"),
}



for i, (metric_name, columns) in enumerate(metrics.items(), 1):
    plt.plot(epochs_df, df[columns[0]], label=f'Transformer {metric_name} Train', marker='o')
    plt.plot(epochs_df, df[columns[1]], label=f'Transformer {metric_name} Validation', linestyle='--', marker='o')
    
    plt.plot(epochs_df2, df2[columns[0]], label=f'Encoder-Decoder {metric_name} Train', marker='s')
    plt.plot(epochs_df2, df2[columns[1]], label=f'Encoder-Decoder {metric_name} Validation', linestyle='--', marker='s')
    
    plt.title(f"{metric_name} Comparison")
    plt.xlabel("Epochs")
    plt.ylabel(metric_name)
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```
